```python
import tensorflow as tf
import pandas as pd
import numpy as np

```

```python
data_csv = pd.read_csv('data.csv', sep = ',', index_col=0)
print(data_csv)
```

Memberi label

```python
bedrooms = data_csv["bedrooms"]
bathrooms = data_csv["bathrooms"]
luas = data_csv["sqft_living"]
floors = data_csv["floors"]
prices = data_csv["price"]
```

Untuk melakukan iterasi

```python
rumah_all =[]
for index in range(len(bedrooms)):
  rumah = []
  rumah.append(bedrooms[index])
  rumah.append(bathrooms[index])
  rumah.append(luas[index])
  rumah.append(floors[index])

  rumah_all.append(rumah)
```

Untuk melakukan iterasi atau pengulangan harus diubah dalam bentuk aaray


```python
print(rumah_all)
rumah_all = np.array(rumah_all)
prices = np.array(prices)
print(rumah_all.shape)
```

Model Machine Learning

```python
model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units= 1, input_shape =[4])

])
```

```python
model.compile(
    loss='mae',
    optimizer='sgd',
    metrics=['accuracy']
)
```

dilakukan perulangan sampe 100

```python
model.fit(
    rumah_all, prices, epochs=100
)
```

melakukan peninjauan terhadap loss nya

Melakukan prediksi

```python
kamar_tidur = int(input("Masukan jumlah kamar tidur yang anda inginkan :"))
kamar_mandi = int(input("Banyak kamar mandi yang diinginkan:"))
luas_rumah = int(input("luas rumah yang diinginkan:"))
jumlah_lantai = int(input("mau ada berapa tingkat ?:"))

prediksi_harga = model.predict([[kamar_tidur, kamar_mandi, luas_rumah, jumlah_lantai]])
print("Rumah impian anda, kira kira harganya {}USD".format(prediksi_harga[0][0]))
```
